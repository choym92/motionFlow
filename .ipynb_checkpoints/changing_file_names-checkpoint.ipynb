`[age]` is an integer from 0 to 116, indicating the age <br>
`[gender]` is either 0 (male) or 1 (female) <br>
`[race]` is an integer from 0 to 4, denoting White, Hispanic, Asian, Indian, and Others (like Hispanic, Latino, Middle Eastern).<br>
`[date&time]` is in the format of yyyymmddHHMMSSFFF, showing the date and time an image was collected to UTKFace<br>

Below code is to adjust **Indian_male** Folder

```
10_0_3.jpg
11_0_3.jpg
12_0_3.jpg
```

````python
import os
import sys

path = './Race/Indian_male/'

for filename in os.listdir(path):
    filename_splitext = os.path.splitext(filename)
    x,y = filename_splitext    
    newname = x.replace('M', '0') + y
#     print(newname)
#     newname = (x[:3] + "M" + "_3" + y)
    os.rename(os.path.join(path, filename), os.path.join(path, newname))
````
